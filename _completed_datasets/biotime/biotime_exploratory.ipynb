### Checking the date data

```python
mdata = data.loc[ (data['MONTH'] > 0) ]
mdata2 = data.loc[ (data['MONTH'] > 0) & (data['DAY'] > 0)]
print("Data:", len(data))
print("MData:", len(mdata))
print("MData2:", len(mdata2))
```

```python
# All rows that have an invalid month value
print(len(data.loc[data['MONTH'] > 12]))
print(len(mdata.loc[mdata['MONTH'] > 12]))
print(len(mdata2.loc[mdata2['MONTH'] > 12]))
```

```python
# All rows that have an invalid month value that could have had their month-day values switched.
len(mdata2.loc[(mdata2['MONTH'] > 12) & (mdata2['DAY'] <= 12)])
```

```python
# All rows with technically valid month values that could also be wrong
at_risk_rows = len(mdata2.loc[(mdata2['MONTH'] <= 12) & (mdata2['DAY'] <= 12)])
print(at_risk_rows)
```

```python
# Remaining rows that are presumably safe
print((len(mdata2) - at_risk_rows)/len(mdata2)* 100, "%")
```

```python
print((len(mdata) - at_risk_rows)/len(mdata)* 100, "%")
```

```python
# Finding the study IDs of the switched rows
mdata2.loc[(mdata2['MONTH'] > 12) & (mdata2['DAY'] <= 12)]['STUDY_ID'].value_counts()
```

```python
len(mdata2.loc[ mdata2['STUDY_ID'].isin([321, 54])])
```

```python
len(data.loc[ data['STUDY_ID'].isin([321, 54])])
```

#### Conclusion:
All invalid month data could have been the result of Americanized date systems, getting the month and the day mixed up.
However, given that this has happened, that means that there are at least `3 Million` rows of data in which the month and day could still have been mixed up, without being able to tell.
