```python
r1 = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=text&starttime=1600-01-01&endtime=1950-01-01")
r2 = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=text&starttime=1950-01-01&endtime=1970-01-01")
r3 = requests.get("https://earthquake.usgs.gov/fdsnws/event/1/query?format=text&starttime=1970-01-01&endtime=1974-01-01")
print(r1, r2, r3)
```

```python
request_url_template = "https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={}-{}-{}&endtime={}-{}-{}"

for year in range(1638, 2020):
    
    sys.stdout.write("\r{}".format(year))
    sys.stdout.flush()
    
    for month in range(1, 13):
        if month == 12:
            end_year = year + 1
            end_month = 1
        else:
            end_year = year
            end_month = month + 1
        
        r = requests.get(request_url_template.format(year, month, 1, end_year, end_month, 1))
        
        if r:
            
            if len(r.json()['features']) == 0:
                continue
            
            else:
                filename = "data/geojson/quakes_{}-{}.json".format(year, month)
                with open(filename, "w") as f:
                    f.write(r.text)
        
        else:
            r1 = requests.get(request_url_template.format(year, month, 1, year, month, 10))
            r2 = requests.get(request_url_template.format(year, month, 10, year, month, 20))
            r3 = requests.get(request_url_template.format(year, month, 20, end_year, end_month, 1))
            
            if r1 and r2 and r3:
                fn1 = "data/geojson/quakes_{}-{}_1.json".format(year, month)
                fn2 = "data/geojson/quakes_{}-{}_2.json".format(year, month)
                fn3 = "data/geojson/quakes_{}-{}_3.json".format(year, month)
                
                with open(fn1, "w") as f1:
                    f1.write(r1.text)
                with open(fn2, "w") as f2:
                    f2.write(r2.text)
                with open(fn3, "w") as f3:
                    f3.write(r3.text)
                    
            else:
                logfile = "data/geojson_logs.txt"
                with open(logfile, "a+") as f:
                    f.write("{}_{}\n".format(year, month))

sys.stdout.write("\rComplete.")
sys.stdout.flush()                    
                    
```
