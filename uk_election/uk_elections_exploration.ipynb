```python
df19 = pd.read_csv("HoC-GE2019-results-by-constituency.csv")
```

```python
df19.columns
```

```python
voters = df19[['electorate','valid_votes','invalid_votes']].sum()
party_votes = df19[['con','lab','ld','brexit','green','snp','pc','dup','sf','sdlp','uup','alliance','other']].sum()
party_seats = df19['first_party'].value_counts()
```

```python
voter_turnout = voters['valid_votes'] / voters['electorate'] * 100
invalid_votes = voters['invalid_votes'] / voters['valid_votes'] * 100
tory_vote_share = party_votes['con'] / party_votes.sum() * 100
tory_seat_share = party_seats['Con'] / party_seats.sum() * 100
print("Voter Turnout: {}%".format(round(voter_turnout,2)))
print("Invalid Votes: {}%".format(round(invalid_votes,2)))
print("Conservative Vote Share: {}%".format(round(tory_vote_share,2)))
print("Conservative Seat Share: {}%".format(round(tory_seat_share,2)))
```

```python
party_seat_shares = round(party_seats / party_seats.sum() * 100,2)
party_vote_shares = round(party_votes / party_votes.sum() * 100,2)
```

```python
party_seat_shares.index = party_seat_shares.index.str.lower()
```

```python
party_seat_shares
```

```python
party_vote_shares
```

```python
party_shares = pd.concat([party_vote_shares, party_seat_shares], axis=1)
party_shares.columns=["Votes","Seats"]
```

```python
party_shares.sort_values(by=['Votes'],ascending=False,inplace=True)
```

```python
party_shares['v/s'] = party_shares['Seats']/party_shares['Votes']
```

```python
party_shares.sort_values(by=['v/s'])
```
