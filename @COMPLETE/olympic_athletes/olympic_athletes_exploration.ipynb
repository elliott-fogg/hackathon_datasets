```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
df.head()
```

```python
sorted(df['NOC'].unique())
```

```python
df['gold'] = np.where(df['Medal']=="Gold", 1, 0)
df['silver'] = np.where(df['Medal']=="Silver", 1, 0)
df['bronze'] = np.where(df['Medal']=="Bronze", 1, 0)
```

```python
medal_bool = df[["NOC","gold","silver","bronze"]]
```

```python
medals = medal_bool.groupby("NOC").sum().sort_values(by=["NOC"])
medals
```

```python
name_map = pd.read_csv("noc_regions.csv")
name_map.set_index("NOC",inplace=True)
name_map
```

```python
country_medals = pd.concat([name_map, medals], axis=1, sort=True)
country_medals
```

```python
country_medals.drop(columns=["notes"]).sort_values(by=["gold","silver"], ascending=False)
```
